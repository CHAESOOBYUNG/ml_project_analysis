```python
import pandas as pd

# CSV 파일 경로 지정
file1 = '..//_ment_death_per_1000k.csv'  # 첫 번째 CSV 파일 경로
file2 = '..//_pop_by_area.csv'  # 두 번째 CSV 파일 경로

# 두 파일을 pandas DataFrame으로 읽어들임
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# 첫 번째 행과 첫 번째 열을 분리하여 저장
first_row = df1.iloc[0, :]  # 첫 번째 행
first_column = df1.iloc[:, 0]  # 첫 번째 열

# 첫 번째 행과 첫 번째 열을 제외한 숫자 데이터만 가져옴
df1_numeric = df1.iloc[1:, 1:].apply(pd.to_numeric, errors='coerce')
df2_numeric = df2.iloc[1:, 1:].apply(pd.to_numeric, errors='coerce')

# 같은 행과 열에 있는 값들을 곱한 후 100000으로 나누기
result_numeric = (df1_numeric * df2_numeric) / 100000

# 첫 번째 행과 첫 번째 열의 데이터를 복원
result_numeric.insert(0, df1.columns[0], first_column.values)  # 첫 번째 열 복원
result_numeric.loc[-1] = first_row.values  # 첫 번째 행 복원
result_numeric.index = result_numeric.index + 1  # 인덱스 재정렬
result_numeric = result_numeric.sort_index()  # 인덱스를 다시 오름차순으로 정렬

# 결과를 새로운 CSV 파일로 저장
result_numeric.to_csv('result.csv', index=False)

print("CSV 파일 처리 완료. 결과는 'result.csv'에 저장되었습니다.")
```
